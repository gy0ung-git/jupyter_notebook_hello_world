````python
# This will convert 200 minutes to hours by dividing by 60
minutes = 200
hours = minutes / 60

print(f"{minutes} minutes is equal to {hours} hours.")

# If you want to display the result with 2 decimal places:
print(f"{minutes} minutes is equal to {hours:.2f} hours.")
````

# Introduction to Data Science Concepts

Data science encompasses various concepts and techniques that are essential for analyzing and interpreting data. Here are some key concepts in data science:

- **Data Exploration**: The process of analyzing data sets to summarize their main characteristics.
- **Data Cleaning**: The process of correcting or removing inaccurate records from a dataset.
- **Feature Engineering**: The creation of new features or variables from existing data to improve model performance.
- **Model Training**: The process of teaching a machine learning model using training data.
- **Model Evaluation**: Assessing the performance of a model using metrics such as accuracy, precision, and recall.
- **Data Visualization**: The graphical representation of information and data to communicate insights clearly
